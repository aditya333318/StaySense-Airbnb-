```python
drop_Listings_df_cols = [ "district","name","host_location","host_response_time",
    "host_picture_url","host_experience_years","host_since"
]

Listings_df.drop(columns=drop_Listings_df_cols, inplace=True, errors="ignore")
```

```python
mean = Listings_df[
    [
        "review_scores_rating",
        "review_scores_accuracy",
        "review_scores_cleanliness",
        "review_scores_checkin",
        "review_scores_communication",
        "review_scores_location",
        "review_scores_value",
    ]
].mean()

```

```python
mean
```

```python
review_cols = [
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value",
]

mean = Listings_df[review_cols].mean()

Listings_df[review_cols] = Listings_df[review_cols].fillna(mean)

```

```python
for col in ["host_response_rate", "host_acceptance_rate"]:
    Listings_df[col] = (
        Listings_df[col]
        .astype(str)
        .str.replace("%","", regex=False)
        .replace("nan", None)
        .astype(float)
    )
    Listings_df[col].fillna(Listings_df[col].median(), inplace=True)
```

```python
Listings_df["host_is_superhost"].fillna("f", inplace=True)
```

```python
Listings_df["host_total_listings_count"].fillna(
    Listings_df["host_total_listings_count"].median(),
    inplace=True
)
```

```python
for col in ["host_has_profile_pic", "host_identity_verified"]:
    Listings_df[col].fillna("f", inplace=True)
```

```python
Listings_df["bedrooms"].fillna(Listings_df["bedrooms"].median(), inplace=True)
```
